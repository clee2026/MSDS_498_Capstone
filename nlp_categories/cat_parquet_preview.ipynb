```
# Display sample rows from NLP issue category parquet
# Shows a variety of issue categories and saves a small CSV output

import pandas as pd
from pathlib import Path

# 1. Config
PARQUET_PATH = "/content/nyc311_with_nlp_issue_categories.parquet"
OUTPUT_DIR = "/content/nlp_issue_outputs"
OUTPUT_SAMPLE_CSV = f"{OUTPUT_DIR}/sample_issue_category_preview.csv"

ROWS_PER_CATEGORY = 3
RANDOM_STATE = 42

Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)

# 2. Load parquet
df = pd.read_parquet(PARQUET_PATH)

print("Loaded parquet successfully")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

# 3. Basic validation
required_cols = ["issue_cluster", "issue_category"]

missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

print("\nIssue category counts:")
display(df["issue_category"].value_counts().reset_index().rename(
    columns={"index": "issue_category", "issue_category": "record_count"}
))

# 4. Select useful display columns
preferred_cols = [
    "unique_key",
    "created_date",
    "closed_date",
    "complaint_type",
    "descriptor",
    "borough",
    "agency_name",
    "location_type",
    "issue_cluster",
    "issue_category",
    "clean_text"
]

cols_to_show = [c for c in preferred_cols if c in df.columns]

print("\nColumns used for preview:")
print(cols_to_show)


# 5. Create varied category sample
sample_df = (
    df.groupby("issue_category", group_keys=False)
      .apply(lambda x: x.sample(min(ROWS_PER_CATEGORY, len(x)), random_state=RANDOM_STATE))
      .reset_index(drop=True)
)

sample_df = sample_df[cols_to_show].sort_values(["issue_category", "issue_cluster"])

print("\nSample rows by issue category:")
display(sample_df)


# 6. Save preview sample to CSV
sample_df.to_csv(OUTPUT_SAMPLE_CSV, index=False)

print("\nSaved sample preview CSV to:")
print(OUTPUT_SAMPLE_CSV)


# 7. Optional download
try:
    from google.colab import files
    files.download(OUTPUT_SAMPLE_CSV)
except Exception as e:
    print("Download skipped:", e)
```
